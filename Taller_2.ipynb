```python
#PUNTO D

def en_riesgo(Estudiantes_completos):

    estudiantes_riesgosos = {}
    
    for key, estudiantes in Estudiantes_completos.items():
        if estudiantes["Promedio_actual"] < 3.0:
         estudiantes_riesgosos[estudiantes["Nombre"]] = estudiantes["Promedio_actual"]
    return estudiantes_riesgosos
        
resultado = en_riesgo(Estudiantes_completos)
print(resultado)        
        
```

```python
#PUNTO E




def existe_est(estudiante_1, estudiante_2, estudiante_3, estudiante_4, nombre_apellido):

   # Convertimos el nombre y apellido a minúsculas para hacer la comparación
    nombre_apellido = nombre_apellido.lower()

    # Función para normalizar el texto
    def normalizar_texto(texto):
        return texto.lower().replace("á", "a").replace("é", "e").replace("í", "i").replace("ó", "o").replace("ú", "u")

    # Comprobamos si el nombre y/o apellido se encuentra en alguno de los estudiantes
    for estudiante in [estudiante_1, estudiante_2, estudiante_3, estudiante_4]:
        if normalizar_texto(nombre_apellido) in normalizar_texto(estudiante):
            return True
    
    # Si no se encuentra en ninguno de los estudiantes, devuelve False
    return False

# Ejemplo de uso
estudiante_1 = "Juan Pérez"
estudiante_2 = "María López"
estudiante_3 = "Pedro Gómez"
estudiante_4 = "Ana Martínez"

nombre_apellido = "lopez"
print(existe_est(estudiante_1, estudiante_2, estudiante_3, estudiante_4, nombre_apellido))  

    
    
```
